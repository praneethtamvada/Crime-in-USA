```python
crimedata['hatecrime'] = (crimedata[['Violent_crime_total','Murder_and_Manslaughter','Forcible_rape','Aggravated_assault']].sum(axis=1))
crimedata['propertycrime'] = (crimedata[['Property_crime_total','Robbery','Burglary','Larceny_theft','Motor_vehicle_theft']].sum(axis=1))
```

```python
unemployment = pd.read_csv('unemployment.csv')
```

```python
sumpop = crimedata.groupby('State',as_index=False)['Population'].sum()
unemployment_bystate = sumpop.join(unemployment['percent_pop'])
unemployment_bystate['unemployment_pop']=(unemployment_bystate['Population']*unemployment_bystate['percent_pop'])/100
unemployment_bystate['unemployment_pop']=unemployment_bystate['unemployment_pop'].astype(int)
unemployment_bystate
```

```python
objects = ('AL','AR','AZ','CA','CO','CT','FL','GA','IA','ID','IL','IN','KS','LA','MA','MI','MN','MO','MS','MT','NC','ND','NH','NJ','NV','NY','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VA','WA','WI')
y_pos = np.arange(len(objects))
performance = sumpop
 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('population in numbers')
plt.title('Population of different States')
plt.show()
```

```python
hatecrime_state = crimedata.groupby('State')['hatecrime'].sum()
propertycrime_state = crimedata.groupby('State')['propertycrime'].sum()
```

```python
width = 0.35
p1 = plt.bar(y_pos, hatecrime_state, width, color='b')
p2 = plt.bar(y_pos, propertycrime_state,width, color='r',bottom=hatecrime_state)
plt.xticks(y_pos, objects)
plt.ylabel('Hate and Property Crime  in Numbers')
plt.title('Hate and Property Crime in different States')
plt.legend((p1[0], p2[0]), ('Hate Crime', 'Property Crime'))
plt.show()
```
